### **1-1. 인공지능과 머신러닝, 딥러닝**

**인공지능(Artificial Intelligence;AI)이란?**
*   사람처럼 학습하고 추론할 수 있는 지능을 가진 컴퓨터 시스템을 만드는 기술

**인공지능 태동기**
*   1943년 워런 매컬러(Warren McCulloch)와 월터 피츠(Walter Pitts)가 최초로 뇌의 뉴런 개념을 발표
*   1950년 앨런 튜링(Alan Turing)은 인공지능이 사람과 같은 지능을 가졌는지 테스트할 수 있는 유명한 튜링 테스트를 발표
*   1956년 다트머스 AI 컨퍼런스에서 인공지능에 대한 밝은 전망이 최고조에 도달

**인공지능 황금기**

*   1957년 프랑크 로젠블라트(Frank Rosenblatt)가 퍼셉트론을 발표
*   1959년 데이비드 허블(David Hubel)과 토르스텐 비젤(Torsten Wiesel)이 고양이를 이용하여 시각 피질에 있는 뉴런 기능을 연구

**강인공지능과 약인공지능**

*   강인공지능(Strong AI;인공일반지능;Artificial General Intelligence): 사람과 구분하기 어려운 지능을 가진 컴퓨터 시스템
*   약인공지능(Weak AI): 특정 분야에서 사람의 일을 도와주는 보조 역할만 가능한 인공지능

**머신러닝(Machine Learing)이란?**

*   규칙을 일일이 프로그래밍하지 않아도 자동으로 데이터에서 규칙을 학습하는 알고리즘을 연구하는 분야
*   지능을 구현하기 위한 소프트웨어를 담당하는 핵심 분야
*   사이킷런(scikit-learn)은 머신러닝의 대표적인 라이브러리

**딥러닝(Deep Learning)이란?**

*   머신러닝 알고리즘 중 인공 신경망(Artifical Neural Network)을 기반으로 한 방법
*   딥러닝은 머신러닝에 포함되고, 머신러닝은 인공지능에 포함되는 개념
*   우리나라에서는 2016년 있었던 이세돌과 알파고의 대국 이후 딥러닝에 대한 관심이 증대
*   2022년 ChatGPT가 출시되면서 대중이 인공지능 기술을 직접 활용하는 시대가 도래
*   텐서플로(TensorFlow), 파이토치(PyTorch) 등은 딥러닝의 대표적인 라이브러리

### **1-2. 코랩과 주피터 노트북**

**구글 코랩(Google Colab)**

*   Python 프로그램을 웹 브라우저에서 무료로 실행 및 저장할 수 있는 서비스
*   Google 계정이 있는 누구나 사용 가능
*   코랩을 사용하면 컴퓨터 사양에 관계없이 자유로운 프로그램 실습이 가능
*   코랩 파일을 노트북(Notebook)이라고 정의

**셀(Cell)**

*   코랩에서 실행할 수 있는 최소 단위
*   텍스트 셀과 코드 셀로 구분

**텍스트(Text) 셀**

*   글을 쓰거나 그림을 삽입하는 등 자유로운 편집이 가능한 셀
*   코드 셀과는 달리 실행되지 않기 때문에 자유롭게 사용 가능

**코드(Code) 셀**

*   코드를 입력하여 실행할 수 있는 셀
*   코드 셀을 실행하면 코랩은 구글 클라우드에 있는 컴퓨터 엔진에서 이 코드를 실행한 다음 그 결과를 반환(코랩은 직접 코드를 실행하지 않음)

**노트북(Notebook)**

*   코랩은 대화형 프로그래밍 환경인 주피터(Jupyter)를 커스터마이징한 것
*   노트북은 주피터 프로젝트의 대표 제품 중 하나로, 주피터 노트북(Jupyter Notebook)이라고도 함
*   코랩 노트북은 구글 클라우드의 가상 서버(Virtual Machine)를 이용하는데, 우상단의 RAM, 디스크 버튼을 누르면 상세 정보 확인이 가능
*   구글 계정만 있다면 누구나 코랩 노트북을 이용하여 무료로 가상 서버를 활용 가능
